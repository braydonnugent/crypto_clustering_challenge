```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
stock_data_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", "price_change_percentage_7d",	"price_change_percentage_14d",	"price_change_percentage_30d",	"price_change_percentage_60d",	"price_change_percentage_200d", "price_change_percentage_1y"]])
```

```python
# Create a DataFrame with the scaled data
df_scaled_data = pd.DataFrame(stock_data_scaled, columns=["price_change_percentage_24h", "price_change_percentage_7d",	"price_change_percentage_14d",	"price_change_percentage_30d",	"price_change_percentage_60d",	"price_change_percentage_200d", "price_change_percentage_1y"])

# Copy the crypto names from the original data
df_scaled_data['coin_id'] = df_market_data.index

# Set the coinid column as index
df_scaled_data.set_index(['coin_id'], inplace=True)

# Display sample data
df_scaled_data.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop variable k for the n_clusters
# 2. Fit the model to the data using df_scaled_data
# 3. Append the model.inertia_ to the inertia list

for i in k:
    kmodel = KMeans(n_clusters=i, random_state=1)
    kmodel.fit(df_scaled_data)
    inertia.append(kmodel.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="elbow_data",
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 
Through analysis of this elbow data, it would appear 4 would be an appropriate value for k as the difference between each value from 4 and beyond are less drastic.

One could argue that 5 would also be an appropriate value for k. Whilst there is still only a rather subtle difference between 4 & 5, it could introduce a more precise perspective on the data.

In saying this, it would be more wise to initialise with a k value of 4.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_scaled_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_segmants_4 = model.predict(df_scaled_data)

# Print the resulting array of cluster values.
print(crypto_segmants_4)
```

```python
# Create a copy of the DataFrame
copy_scaled_data = df_scaled_data
```

```python
# Add a new column to the DataFrame with the predicted clusters
copy_scaled_data['crypto_segmants_4'] = crypto_segmants_4

# Display sample data
copy_scaled_data.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

copy_scaled_data.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="crypto_segmants_4",
    hover_cols="coin_id"  
)
```

---

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
transformed_data = pca.fit_transform(df_scaled_data)

# View the first five rows of the DataFrame. 
print(transformed_data[0:5])
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.

explained_variance = pca.explained_variance_ratio_

print("Explained Variance for Each Principal Component:")
for i, variance in enumerate(explained_variance):
    print(f"Principal Component {i+1}: {variance:.2f}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** To understand the total variance these 3 components represent, it would be a matter of summing them together. After doing so, we have a total of 0.89, which ultimately means these principal components altogether capture a total of 89% of the variability within the original dataset.

```python
# Create a new DataFrame with the PCA data.
pca_df = pd.DataFrame(transformed_data, columns=['PC1', 'PC2', 'PC3'])

# Creating a DataFrame with the PCA data...
# Copy the crypto names from the original data
pca_df['coin_id'] = df_scaled_data.index

# Set the coinid column as index
pca_df.set_index(['coin_id'], inplace=True)

# Display sample data
pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `pca_df`, which contains the PCA-transformed data
# 3. Append the model.inertia_ to the inertia list

for i in k:
    kmodel = KMeans(n_clusters=i, random_state=1)
    kmodel.fit(pca_df)
    inertia.append(kmodel.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)
df_elbow_pca
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow_pca.hvplot.line(
    x="k",
    y="inertia",
    title="elbow_data",
    xticks=k
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** After analysing this elbow data, it would appear that the optimal value for k to cluster the data would be 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** There does appear to be a slightly different range between the pca & original data, but ultimately the optimal k value remains the same.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
crypto_segmants_4 = model.predict(pca_df)

# Print the resulting array of cluster values.
print(crypto_segmants_4)
```

```python
# Create a copy of the DataFrame with the PCA data
copy_pca_df = pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
copy_pca_df['crypto_segmants_4'] = crypto_segmants_4

# Display sample data
copy_pca_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

copy_pca_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="crypto_segmants_4",
    hover_cols="coin_id"
)
```
